## Part 1 ‚Äê Exploratory data analysis
The attached logins.json file contains (simulated) timestamps of user logins in a particular geographic location. Aggregate these login counts based on 15minute time intervals, 
and visualize and describe the resulting time series of login counts in ways that best characterize the underlying patterns of the demand. 
Please report/illustrate important features of the demand, such as daily cycles.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('logins.json')
```

```python
df.head()
```

```python
df.info()
```

### 1) Aggregating login counts based on 15 min time intervals

```python
df['count'] =1
```

```python
df_grouped = df.groupby(pd.Grouper(key = 'login_time', freq='15min')).count()
```

```python
df_grouped.head()
```

### 2) Time series

```python
import matplotlib.pyplot as plt
```

```python
df_grouped = df_grouped.reset_index()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,9)
ax.plot('login_time','count',data=df_grouped)
ax.set_title('Time Series of Login Counts 15 Minutes Intervals')
ax.set_xlabel('Time')
ax.set_ylabel('Counts')
plt.show()
```

```python
df_day = df_grouped[(df_grouped['login_time']<'1970-01-03') & (df_grouped['login_time']>='1970-01-02')]
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,9)
ax.plot('login_time','count',data=df_day)
ax.set_title('Time Series of Login Counts 15 Minutes Intervals 1970-01-02')
ax.set_xlabel('Time')
ax.set_ylabel('Counts')
plt.show()
```

```python
df_week = df_grouped[df_grouped['login_time']<'1970-01-08']
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,9)
ax.plot('login_time','count',data=df_week)
ax.set_title('Time Series of Login Counts 15 Minutes Intervals First Week')
ax.set_xlabel('Time')
ax.set_ylabel('Counts')
plt.show()
```

```python
df_month = df_grouped[df_grouped['login_time']<'1970-02-01']
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,9)
ax.plot('login_time','count',data=df_month)
ax.set_title('Time Series of Login Counts 15 Minutes Intervals First Month')
ax.set_xlabel('Time')
ax.set_ylabel('Counts')
plt.show()
```

I couldn't find a noticeable trend probabaly due to the lack of data only lasting few months. \
From the overall time series, there seems to be a weekly pattern with a gradual increase then a drop. \
From the first week graph, I could see that the higest login counts was reached on 1970-01-04, which was Sunday. The counts then significantly dropped following Monday. \
From the first month graph, I could verify the weekely pattern with the Sunday reaching the highest every week. \
Although every day seems different, there seem to be a pattern of increased counts in the morning, noon, and evening (W shaped counts).
